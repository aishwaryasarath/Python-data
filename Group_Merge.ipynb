# Difference cases while loading data


1.   pd.read_csv("example.csv)
     **name\tgender\tage\toaths** <- If columns are in this format
     Then do pd.read_csv("example.csv, delim_whitespace=True)
2.   If columns have an unnamed index
     pd.read_csv("example.csv, index_col=0)
3.   If delimiter is | and header is not identified
4.   pd.read_csv("example.csv", sep = "||, comment="#")



```python
import pandas as pd
#df = pd.read_csv("/content/sample_data/cost-of-living.csv", index_col=0)
df = pd.read_csv("/content/sample_data/cost-of-living.csv", index_col=0).T.reset_index()
df.head(15)
```

```python
df2 = df.rename(columns ={'index':'location'})
df.head()
```

```python
df2[["City", "Country"]] = df2.location.str.split(", ", n=1, expand=True)
df2.head()
```

```python
df2.Country.unique()
```

```python
df2.Country = df2.Country.apply(lambda x: x if "," not in x else x.split(", ")[-1])
df2.Country.unique()
```

```python
df2.Country.value_counts()
```

```python
most_cities = df2.Country.value_counts().index[0]
most_cities
```

```python
df3 = df2[df2.Country == most_cities]
df3.head(5)
```

```python
df.info()
```

```python
col = "Apartment (1 bedroom) Outside of Centre"
df4 = df3.sort_values(col, ascending=False)
df4[["location", col]]
```
